```python
import numpy as np
import pandas as pd
```

```python
df = pd.read_csv('scores_data.csv')

df.head()
```

```python
df.info()
```

```python
df.columns
```

```python
# Remove spaces from columns

df.columns = df.columns.str.replace(' ', '')
```

```python
df.columns
```

```python
df['New Score'] = df['Score'].apply(lambda x : (int(x[0])/int(x[4]))*100)
```

```python
# pd.to_numeric(df['Score'])
df['New Score']
```

```python
df['New Score'].plot(kind='hist')
```

```python
df['New Score'].plot(kind='kde')
```

# Batch column

```python
import seaborn as sns
```

```python
df['Batch'].value_counts()
```

```python
print(df[df['Batch'] == 'AI_ELITE_7']['New Score'].mean())

print(df[df['Batch'] == 'AI_ELITE_7']['New Score'].median())
```

```python
print(df[df['Batch'] == 'AI_ELITE_6']['New Score'].mean())

print(df[df['Batch'] == 'AI_ELITE_6']['New Score'].median())
```

```python
print(df[df['Batch'] == 'AI_ELITE_4']['New Score'].mean())

print(df[df['Batch'] == 'AI_ELITE_4']['New Score'].median())
```

```python
sns.countplot(x=df["Batch"])
```

```python
df.boxplot(by ='Batch', column =['New Score'], grid = False)
```

```python
df.plot.scatter(x = 'Batch', y = 'New Score', s = 100);
```

```python
sns.catplot(data=df, x="Batch", y="New Score", hue="Batch", kind="swarm")
```
